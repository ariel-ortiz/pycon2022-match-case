# `traduce`

```python
def traduce_v1(palabra):
    if palabra == 'pollito':
        return 'chicken'
    elif palabra == 'gallina':
        return 'hen'
    elif palabra == 'lápiz':
        return 'pencil'
    elif palabra == 'pluma':
        return 'pen'
    else: 
        return 'sepa la bola'
```

```python
traduce_v1('pollito')
```

```python
def traduce_v2(palabra):
    match palabra:
        case 'pollito':
            return 'chicken'
        case 'gallina':
            return 'hen'
        case 'lápiz':
            return 'pencil'
        case 'pluma':
            return 'pen'
        case _:
            return 'sepa la bola'
```

```python
traduce_v2('pollito')
```

```python
DICC = {
    'pollito': 'chicken',
    'gallina': 'hen',
    'lápiz': 'pencil',
    'pluma': 'pen',
}

def traduce_v3(palabra):
    return DICC.get(palabra, 'sepa la bola')
```

```python
traduce_v3('pollito')
```

# `dias_en_mes`

```python
def dias_en_mes_v1(mes):
    if mes in [11, 4, 6, 9]:
        return 30
    elif mes == 2:
        return 28
    else:
        return 31
```

```python
dias_en_mes_v1(3)
```

```python
def dias_en_mes_v2(mes):
    match mes:
        case 11 | 4 | 6 | 9:
            return 30
        case 2:
            return 28
        case _:
            return 31
```

```python
dias_en_mes_v2(2)
```

# `rango_carta`

```python
def rango_carta(carta):
    match carta:
        case ['A', _]:
            return 14
        case ['K', _]:
            return 13
        case ['Q', _]:
            return 12
        case ['J', _]:
            return 11
        case [x, _] if isinstance(x, int) and 2 <= x <= 10:
            return x
        case _:
            raise ValueError(f'Carta inválida: {carta}')
```

```python
rango_carta(('J', 'pica'))
```

```python
baraja = [('Q', 'corazon'), 
          (5, 'trebol'), 
          ('J', 'pica'),
          (7, 'diamante')]
```

```python
sorted(baraja, key=rango_carta)
```

# `fizz_buzz`

```python
def fizz_buzz():
    for i in range(1, 101):
        match (i % 3, i % 5):
            case (0, 0):
                print('FizzBuzz', end=', ')
            case (0, _):
                print('Fizz', end=', ')
            case (_, 0):
                print('Buzz', end=', ')
            case _:
                print(i, end=', ')
```

```python
def es_cuadro_magico(tablero):
    lista = tablero[0] + tablero[1] + tablero[2]
    lista.sort()
    if (len(lista) != 9
            or len(set(lista)) != 9
            or lista[0] != 1
            or lista[-1] != 9): 
        return False
    match tablero:
        case [[a, b, c],
              [_, _, _],
              [_, _, _]] if a + b + c != 15:
            return False
        case [[_, _, _],
              [a, b, c],
              [_, _, _]] if a + b + c != 15:
            return False
        case [[_, _, _],
              [_, _, _],
              [a, b, c]] if a + b + c != 15:
            return False
        case [[a, _, _],
              [b, _, _],
              [c, _, _]] if a + b + c != 15:
            return False
        case [[_, a, _],
              [_, b, _],
              [_, c, _]] if a + b + c != 15:
            return False
        case [[_, _, a],
              [_, _, b],
              [_, _, c]] if a + b + c != 15:
            return False
        case [[a, _, _],
              [_, b, _],
              [_, _, c]] if a + b + c != 15:
            return False
        case [[_, _, c],
              [_, b, _],
              [a, _, _]] if a + b + c != 15:
            return False
        case [[_, _, _],
              [_, _, _],
              [_, _, _]]:
            return True
        case _:
            return False
```

```python
es_cuadro_magico([[8, 1, 6],
                  [3, 5, 7],
                  [4, 9, 2]])
```

```python
es_cuadro_magico([[1], 
                  [2, 3, 4, 5, 6],
                  [7, 8, 9]])
```

# `aplana`

```python
def aplana(lista):
    match lista:
        case []:
            return []
        case [cabeza, *cola] if isinstance(cabeza, list):
            return aplana(cabeza) + aplana(cola)
        case [cabeza, *cola]:
            return [cabeza] + aplana(cola)
```

```python
aplana([[[[1]]], [[2], [3, [4], [5, 6]]]])
```
